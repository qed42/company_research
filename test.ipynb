```python
import requests

response = requests.post(
    "http://localhost:8000/research_company",
    json={"company_input": "NSE: HYUNDAI"}
)
company_data = response.json()
```

```python
from IPython.display import display, Markdown
from datetime import datetime
import pytz

def display_company_research(company_data):
    """
    Formats and displays company research data in a Jupyter notebook.
    
    Args:
        company_data: JSON response from the company research API
    """
    # Header
    md_content = f"""# {company_data['input']} Research Report
*Generated on {datetime.fromisoformat(company_data['timestamp']).astimezone(pytz.UTC).strftime('%B %d, %Y %H:%M:%S')} UTC*

**Company Information:** {company_data['resolved_info']}

---
"""

    # Sections ordering and titles
    section_order = [
        ('introduction', '## Introduction'),
        ('founded_year', '### Founded'),
        ('managing_director', '### Leadership'),
        ('company_overview', '## Company Overview'),
        ('business_segments', '## Business Segments'),
        ('leadership', '## Leadership Structure'),
        ('financial_performance', '## Financial Performance'),
        ('business_segment_deep_dive', '## Business Segment Analysis'),
        ('recent_developments', '## Recent Developments'),
        ('industry_outlook', '## Industry Outlook')
    ]

    # Add each section
    for section_key, section_title in section_order:
        if section_key in company_data['sections']:
            content = company_data['sections'][section_key].strip()
            if content and content.lower() != "information not available.":
                md_content += f"\n{section_title}\n{content}\n"

    # Display the formatted content
    display(Markdown(md_content))
```

```python
display_company_research(company_data)
```
